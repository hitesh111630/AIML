```python
# record_speech.py
import sounddevice as sd
from scipy.io.wavfile import write

fs = 16000  # sample rate
seconds = 10# record duration

print("🎙️ Speak now... Recording for 10 seconds...")
audio = sd.rec(int(seconds * fs), samplerate=fs, channels=1)
sd.wait()
write("myvoice.wav", fs, audio)
print("✅ Recording saved as myvoice.wav")
```

```python
# analyze_speech.py
import librosa
import numpy as np

y, sr = librosa.load("myvoice.wav", sr=16000)

# Duration & energy
duration = librosa.get_duration(y=y, sr=sr)
energy = np.mean(librosa.feature.rms(y=y))
zcr = np.mean(librosa.feature.zero_crossing_rate(y))

# MFCCs (voice tone)
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
mfcc_mean = np.mean(mfcc, axis=1)

print("🎤 Speech Analysis Report:")
print(f"Duration: {duration:.2f} sec")
print(f"Average Energy: {energy:.4f}")
print(f"Zero Crossing Rate: {zcr:.4f}")
print(f"MFCC Mean: {np.round(mfcc_mean, 2)}")
```

```python
# detect_dementia.py
import numpy as np
import joblib

# Assume analysis values (same order as above)
duration = 5.0
energy = 0.05
zcr = 0.1
mfcc_mean = np.random.rand(13)

features = np.concatenate(([duration, energy, zcr], mfcc_mean))
features = features.reshape(1, -1)

# Load trained model (we’ll train later)
try:
    model = joblib.load("dementia_model.pkl")
    pred = model.predict(features)[0]
    if pred == 1:
        print("⚠️ Dementia-like Speech Pattern Detected")
    else:
        print("✅ Normal Speech Pattern")
except:
    print("🧩 Model not trained yet! We'll add this in next step.")
```
