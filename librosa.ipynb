```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# 1. Get the file path to a single included audio example
filename = 'Recordingwav.wav'

# 2. Load the audio as the "slower" version
x_1, sr = librosa.load(filename)

# 3. Create a "faster" version by time-stretching the first one
#    A rate > 1.0 speeds up the audio. Let's make it 20% faster.
x_2 = librosa.effects.time_stretch(x_1, rate=1.2)

display(Audio(data=x_2, rate=sr))
# 4. Now, plot both waveforms
fig, ax = plt.subplots(nrows=2, sharex=True, sharey=True, figsize=(10, 6))


librosa.display.waveshow(x_1, sr=sr, ax=ax[0])
ax[0].set(title='Original Version (x_1)')
ax[0].label_outer()

librosa.display.waveshow(x_2, sr=sr, ax=ax[1])
ax[1].set(title='Faster Version (x_2) at 1.2x Speed')

plt.show()
```
