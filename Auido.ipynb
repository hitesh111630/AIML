```python
# Beat tracking example
import librosa
import numpy as np # Import numpy to calculate the mean

# 1. Get the file path to an included audio example
filename = 


# 2. Load the audio as a waveform `y`
#    Store the sampling rate as `sr`
y, sr = librosa.load(filename)

# 3. Run the default beat tracker
#    'tempo' is now an array of tempo values
tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr,duration=5)

# --- FIX ---
# Calculate the average tempo from the tempo array
average_tempo = np.mean(tempo)

print(f'Estimated average tempo: {average_tempo:.2f} beats per minute')

# 4. Convert the frame indices of beat events into timestamps
beat_times = librosa.frames_to_time(beat_frames, sr=sr)

# You can also print the first few beat timestamps
print('First 5 beat times (s):', beat_times[:5])
```

```python
from pydub import AudioSegment
import os

# Define your input and output filenames
input_file = "Recording.m4a"
output_file = "Recording.wav"

try:
    # Check if the input file exists
    if not os.path.exists(input_file):
        raise FileNotFoundError(f"The input file '{input_file}' was not found.")

    # Load the M4A file and export it as WAV
    print(f"Loading '{input_file}'...")
    audio = AudioSegment.from_file(input_file, format="m4a")
    
    print(f"Converting and saving to '{output_file}'...")
    audio.export(output_file, format="wav")
    
    print("Conversion successful!")

except FileNotFoundError as e:
    print(f"Error: {e}")
except Exception as e:
    # This error often means FFmpeg is not installed or not in your PATH
    print(f"An error occurred: {e}")
    print("Please ensure FFmpeg is installed and accessible in your system's PATH.")
```
